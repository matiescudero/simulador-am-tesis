```python
import osmnx as ox
import networkx as nx
import pandas as pd
import geopandas as gpd
from shapely.geometry import shape
from shapely.wkt import loads
```

```python
# Se pega el string WKT generado
wkt_string = "POLYGON((-70.58558640323943 -33.445877690995374,-70.50524887882537 -33.445877690995374,-70.50524887882537 -33.506588243585746,-70.58558640323943 -33.506588243585746,-70.58558640323943 -33.445877690995374))"

# Convierte el string WKT a un objeto de polígono usando shapely
polygon = loads(wkt_string)

# Descarga el grafo peatonal dentro del polígono
G = ox.graph_from_polygon(polygon, network_type='walk')
ox.plot_graph(G)
```

```python
ox.save_graphml(G, filepath="../data/raw/red_peatonal_penalolen.graphml")
```
